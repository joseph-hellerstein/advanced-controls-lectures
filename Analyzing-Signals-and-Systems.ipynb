## Antimony Models

```python
INITIAL_CONDITIONS_MDL = """
J1: S1 -> S2; k1*S1
J2: S2 -> S3; k2*S2

k1 = 2
k2 = 0.5
S1 = 10
S2 = 0
S3 = 0
"""
rr = te.loada(INITIAL_CONDITIONS_MDL)
data = rr.simulate(0, 10)
rr.plot(data)
```

```python
IMPULSE_RESPONSE_MDL = """
J1: $S1 -> S2; k1*$S1
J2: S2 -> S3; k2*S2

k1 = 2
k2 = 0.5
$S1 = 10
S2 = 0
S3 = 0

at (time > 1): $S1 = 0
"""
rr = te.loada(IMPULSE_RESPONSE_MDL)
data = rr.simulate(0, 10, 1000)
rr.plot(data)
```

```python
IMPULSE_RESPONSE_MDL = """
J1: $S1 -> S2; k1*$S1
J2: S2 -> S3; k2*S2

k1 = 2
k2 = 0.5
$S1 = 1000
S2 = 0
S3 = 0

at (time > 0.01): $S1 = 0
"""
rr = te.loada(IMPULSE_RESPONSE_MDL)
data = rr.simulate(0, 10, 1000)
rr.plot(data)
```

```python
STEP_RESPONSE_MDL = """
J1: $S1 -> S2; k1*$S1
J2: S2 -> S3; k2*S2

k1 = 2
k2 = 0.5
$$S1 = 10
S2 = 0
S3 = 0

"""
rr = te.loada(STEP_RESPONSE_MDL)
data = rr.simulate(0, 10, 1000)
rr.plot(data)
```
