# Sequential (linear) Pathways

$S_1 \rightarrow S_2 \rightarrow S_3$

```python
sequential_model = """

S1 -> S2; k1*S1
S2 -> S3; k2*S2

S1 = 10
S2 = 0
S3 = 0

k1 =1
k2 = 1

"""
rr = te.loada(sequential_model)
```

```python
rr.reset()
data = rr.simulate()
rr.plot(data)
```

```python
ctlsb = ctl.ControlSBML("https://www.ebi.ac.uk/biomodels/model/download/BIOMD0000000823.2?filename=Varusai2018.xml")
```

```python
ctlsb.roadrunner.reset()
data = ctlsb.roadrunner.simulate()
_ = plt.plot(data[:, 0], data[:, 1:])
plt.legend(data.colnames[1:])
```
