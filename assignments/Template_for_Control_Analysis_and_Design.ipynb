# System Identification

System identifical is about constructing a transfer function that approximates the effect of the control input on the measured output. You will construct a transfer function $G(s)$ that is the ratio of the Laplace transfer of the measured output ($Y(s)$) to the Laplae transform of the control input ($U(s)$). You will compare these predictions with those obtained from simulation.

1. How well does your fitted transfer function predict the behavior of the system?
1. Assess the following:
    1. How much does the model improved if you increase the order of the transfer function.
    1. How much does the model degrade if you decrease the order of the transfer function.
1. From your transfer function,
    1. What are the poles?
    1. What do the poles tell you about oscillations?
    1. What do the poles tell you about BIBO (bounded input, bounded output) stability?
